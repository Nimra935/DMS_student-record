```python
#Final Python project file


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
data = pd.read_csv('/content/modified_real_data.csv')

# Cleaning the dataset
data = data.dropna(how='all')  # Drop rows where all elements are NaN

# Function to display menu
def display_menu():
    print("\nChoose an analysis option:")
    print("1. Search for a student")
    print("2. Visualize grade distribution")
    print("3. Show CGPA distribution")
    print("4. Show students by department")
    print("5. Show CGPA vs. Batch scatter plot")
    print("6. Visualize enrollment type")
    print("7. Visualize CGPA vs Batch (Boxplot)")
    print("8. View class statistics")
    print("9. clustering_analysis")
    print("10. Correlation analysis")
    print("11. Time series analysis")
    print("12. Cluster analysis")
    print("13. Predictive recommendations")
    print("14. Plot histograms (CGPA)")
    print("15. Plot boxplots (CGPA)")
    print("16. Exit")

# 1. Search for students by name or registration number
def search_student():

    query = input("Enter student name or registration number: ").strip()
    if query.isdigit():

        result = data[data['Registration No'].astype(str) == query]
    else:
        result = data[data['Name'].str.contains(query, case=False, na=False)]
    if not result.empty:
        print(result)
    else:
        print("No student found with that name or registration number.")

# 2. Visualize grade distribution
def visualize_grade_distribution():
    sns.countplot(data=data, x='Grade')
    plt.title('Grade Distribution')
    plt.xlabel('Grade')
    plt.ylabel('Count')
    plt.show()

# 3. Show CGPA distribution
def show_cgpa_distribution():
    sns.histplot(data=data, x='CGPA', kde=True)
    plt.title('CGPA Distribution')
    plt.xlabel('CGPA')
    plt.ylabel('Frequency')
    plt.show()

# 4. Show students by department
def students_by_department():
    data['Department'].value_counts().plot(kind='bar')
    plt.title('Students by Department')
    plt.xlabel('Department')
    plt.ylabel('Count')
    plt.show()

# 5. Show CGPA vs. Batch scatter plot
def cgpa_vs_batch():
    sns.scatterplot(data=data, x='Batch', y='CGPA')
    plt.title('CGPA vs. Batch')
    plt.xlabel('Batch')
    plt.ylabel('CGPA')
    plt.xticks(rotation=45)
    plt.show()

# 6. Visualize enrollment type
def visualize_enrollment_type():
    sns.countplot(data=data, x='Enrollment Type')
    plt.title('Enrollment Type Distribution')
    plt.xlabel('Enrollment Type')
    plt.ylabel('Count')
    plt.show()

# 7. Visualize CGPA vs Batch (Boxplot)
def cgpa_vs_batch_boxplot():
    sns.boxplot(data=data, x='Batch', y='CGPA')
    plt.title('CGPA vs. Batch')
    plt.xlabel('Batch')
    plt.ylabel('CGPA')
    plt.xticks(rotation=45)
    plt.show()

# 8. View class statistics
def class_statistics():
    total_students = len(data)
    students_with_3_cgpa = len(data[data['CGPA'] == 3])
    students_below_3_cgpa = len(data[data['CGPA'] < 3])
    #students_in_thesis = len(data[data['Thesis Phase'] == 'Yes'])
    #students_in_thesis = len(data[data['Thesis Phase'] == 'Yes'])  # Assuming 'Thesis Phase' column

    print("Total number of students:", total_students)
    print("Number of students with exactly 3 CGPA:", students_with_3_cgpa)
    print("Number of students with CGPA less than 3:", students_below_3_cgpa)
   # print("Number of students in thesis phase:", students_in_thesis)

# 9. Perform predictive analytics
def predictive_analytics():
    features = ['Session', 'CGPA']
    df = data.dropna(subset=features)
    X = df[['Session']]
    y = df['CGPA']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    print("Mean Squared Error:", mean_squared_error(y_test, predictions))
    print("R-squared:", r2_score(y_test, predictions))

# 10.# Advanced Analytics Features
def correlation_analysis():
    print("\n### Correlation Analysis ###")
    numeric_df = df.select_dtypes(include=['float64', 'int64'])
    corr = numeric_df.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix')
    plt.show()

# 11. Time series analysis
def time_series_analysis():
    data['Year'] = data['Session']
    yearly_data = data.groupby('Year')['CGPA'].mean()
    yearly_data.plot(kind='line')
    plt.title('Yearly CGPA Trend')
    plt.xlabel('Year')
    plt.ylabel('Average CGPA')
    plt.show()
# 12. Cluster analysis
def cluster_analysis():
    df = data[['CGPA', 'Session']].dropna()
    kmeans = KMeans(n_clusters=3)
    clusters = kmeans.fit_predict(df)
    plt.scatter(df['Session'], df['CGPA'], c=clusters, cmap='viridis')
    plt.title('Cluster Analysis')
    plt.xlabel('Session')
    plt.ylabel('CGPA')
    plt.show()

# 13. Predictive recommendations
def predictive_recommendations():
    cgpa = float(input("Enter CGPA to get recommendations: "))
    if cgpa > 3.5:
        print("Eligible for advanced courses or thesis options.")
    elif cgpa >= 2.0:
        print("Consider improvement strategies to achieve higher milestones.")
    else:
        print("Seek academic counseling for improvement.")

# 14. Plot histograms
def plot_histograms():
    sns.histplot(data=data, x='CGPA', kde=True)
    plt.title('CGPA Histogram')
    plt.show()

# 15. Plot boxplots
def plot_boxplots():
    sns.boxplot(data=data, y='CGPA')
    plt.title('CGPA Boxplot')
    plt.show()

# Main program loop
while True:
    display_menu()
    choice = input("Enter your choice: ")

    if choice == '1':
        search_student()
    elif choice == '2':
        visualize_grade_distribution()
    elif choice == '3':
        show_cgpa_distribution()
    elif choice == '4':
        students_by_department()
    elif choice == '5':
        cgpa_vs_batch()
    elif choice == '6':
        visualize_enrollment_type()
    elif choice == '7':
        cgpa_vs_batch_boxplot()
    elif choice == '8':
        class_statistics()
    elif choice == '9':
        predictive_analytics()
    elif choice == '10':
        correlation_analysis()
    elif choice == '11':
        time_series_analysis()
    elif choice == '12':
        cluster_analysis()
    elif choice == '13':
        predictive_recommendations()
    elif choice == '14':
        plot_histograms()
    elif choice == '15':
        plot_boxplots()
    elif choice == '16':
        print("Exiting program.")
        break
    else:
        print("Invalid choice. Please try again.")
```
